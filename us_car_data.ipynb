```python
# Import the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Import the dataset
path='/content/drive/MyDrive/Data Science/US Car data (EV)/us_car_data.csv'
df=pd.read_csv(path)
df.head()
```

```python
# Let look at the dataset for null values
df.isnull().sum()
```

```python
# Looks like the data has 0 null values.
# Let look at the datatypes
df.info()
```

```python
# Since the data does not have any null values lets start with univariate analysis
sns.countplot(x='Make', hue='Make',data=df)
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
# Clearly Tesla is the leading brand of EV, with over 60000 cars sold, followed by Nissan and BMW.
```

```python
# City wise countplot for EV's sold in top 15 cities
sns.countplot(y='City', hue='City', data=df, order=df['City'].value_counts().iloc[:15].index)
plt.grid()
plt.show()
```

```python
# Seattle is having the highest sale of EV's out of all, followed by Bellevue and Redmond
```

```python
# County wise sale of EV's in top 15 counties
sns.countplot(y='County', hue='County', data=df, order=df['County'].value_counts().iloc[:15].index)
plt.grid()
plt.show()
```

```python
# Kings has the highest sale of EV's with over 70000 sales
```

```python
# Countplot of different Electric vehicle types
sns.countplot(x='Electric Vehicle Type', hue='Electric Vehicle Type', data=df)
plt.grid()
plt.show()
```

```python
# Battery electric vehicle sales are more than double than Plug-in Hybrid electric vehicles.
```

```python
# Lets look at year on year sale of top 5 manufacturers
plt.figure(figsize=(10,5))
sns.countplot(x='Make', hue='Model Year', data=df, order=df['Make'].value_counts().iloc[:5].index)
plt.grid()
plt.show()
```
